# Clustering - Univariate, Bivariate, Multivariate

```python
clustering1 = KMeans(n_clusters = 3)
```

```python
clustering1.fit(df[['Annual Income (k$)']])
```

```python
clustering1.labels_
```

```python
df['Income Cluster'] = clustering1.labels_
df.head()
```

```python
df['Income Cluster'].value_counts()
```

```python
clustering1.inertia_ #clustering1 score
```

```python
inertia_scores =[]
for i in range (1,20):
    kmeans=KMeans(n_clusters = i)
    kmeans.fit(df[['Annual Income (k$)']])
    inertia_scores.append(kmeans.inertia_)
    print(i, kmeans.inertia_)
```

```python
plt.plot(range(1,20),inertia_scores)
plt.xticks(range(1,20))
#Elbow start at 3, change number of clusters to 3
```

```python
df.groupby('Income Cluster')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean()
```

# Bivariate Clustering

```python
clustering2 = KMeans(n_clusters = 5)
clustering2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
clustering2.labels_
df['Spending and Income Cluster'] = clustering2.labels_
df.head()
```

```python
inertia_scores2 =[]
for i in range (1,20):
    kmeans2=KMeans(n_clusters = i)
    kmeans2.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
    inertia_scores2.append(kmeans2.inertia_)
print(inertia_scores2)
plt.plot(range(1,20),inertia_scores2)
```

Insights: Inertia measures the sum of squared distances between data points and their closest cluster center. Lower inertia generally indicates better clustering, but too many clusters can lead to overfitting. Here we can see the optimal k is around 5. 

```python
centers =pd.DataFrame(clustering2.cluster_centers_)
centers.columns=['x','y']
```

```python
plt.figure(figsize =(10,8))
plt.scatter(x=centers['x'],y=centers['y'],s  = 100, c='black',marker ='*')
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)',hue = 'Spending and Income Cluster',palette = 'tab10')
plt.savefig('clustering_biavariate.png')
```

Clustering with k=5 does a pretty good job at circling similar customer groups.
